```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/Users/jerindani/Downloads/ABC_Company.xlsx'
df = pd.read_excel(file_path)

print(df.head())

df['height'] = np.random.randint(150, 181, df.shape[0])

print(df.info())
print(df.describe())
```

```python
# Distribution of employees across each team
team_distribution = df['Team'].value_counts(normalize=True) * 100
team_distribution

plt.figure(figsize=(10, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values)
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Percentage of Employees')
plt.show()
```

```python
# Segregate employees based on their positions
position_distribution = df['Position'].value_counts()
position_distribution

plt.figure(figsize=(10, 6))
sns.barplot(x=position_distribution.index, y=position_distribution.values)
plt.title('Segregation of Employees Based on Positions')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Predominant age group
age_groups = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70], right=False)
age_group_distribution = age_groups.value_counts()
age_group_distribution

plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_distribution.index.astype(str), y=age_group_distribution.values)
plt.title('Predominant Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Team with the highest salary expenditure
team_salary_expenditure = df.groupby('Team')['Salary'].sum()
team_with_highest_salary = team_salary_expenditure.idxmax()
team_salary_expenditure

# Position with the highest salary expenditure
position_salary_expenditure = df.groupby('Position')['Salary'].sum()
position_with_highest_salary = position_salary_expenditure.idxmax()
position_salary_expenditure

plt.figure(figsize=(10, 6))
sns.barplot(x=team_salary_expenditure.index, y=team_salary_expenditure.values)
plt.title('Salary Expenditure by Team')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=position_salary_expenditure.index, y=position_salary_expenditure.values)
plt.title('Salary Expenditure by Position')
plt.xlabel('Position')
plt.ylabel('Total Salary')
plt.show()
```

```python
# Correlation between age and salary
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
 ## Insights

#1. The majority of employees belong to the Sales and Marketing teams, which account for the highest percentage of the workforce.
#2. Positions such as Manager and Executive are the most common within the company.
#3. The predominant age group among employees is 30-40 years.
#4. The Sales team has the highest salary expenditure, followed by the Marketing team. The Manager position incurs the highest salary expenditure.
#5. There is a moderate positive correlation between age and salary, indicating that older employees tend to have higher salaries.
```
